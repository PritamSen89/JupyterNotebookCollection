### Random Data Distribution

```python
import numpy as np
from numpy import random
```

```python
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))

print(x)

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))

print(x)
```

```python
# shuffle
# shuffle() method makes changes to the original array.
arr = np.array([1, 2, 3, 4, 5])

random.shuffle(arr)

print(arr)

arr = np.array([1, 2, 3, 4, 5])

print(random.permutation(arr))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.distplot([0, 1, 2, 3, 4, 5])

plt.show()

sns.distplot([0, 1, 2, 3, 4, 5], hist=False)

plt.show()
```

```python
# Normal (Gaussian) Distribution
x = random.normal(size=(2, 3))

print(x)

x = random.normal(loc=1, scale=2, size=(2, 3)) # mean 1, std 2

print(x)


sns.distplot(random.normal(size=1000), hist=False)

plt.show()
```

```python
# Binomial Distribution
# Binomial Distribution is a Discrete Distribution.

x = random.binomial(n=10, p=0.5, size=10) # n : number of trials, p : probability of occurence of each trial 

print(x)

sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)

plt.show()
```

```python
# Poisson Distribution
# Poisson Distribution is a Discrete Distribution.
#It estimates how many times an event can happen in a specified time. e.g. If someone eats twice a day what is probability he will eat thrice?

x = random.poisson(lam=2, size=10)

print(x)

sns.distplot(random.poisson(lam=2, size=1000), kde=False) # lam - rate or known number of occurences e.g. 2 

plt.show()
```

```python
# Uniform Distribution
# Used to describe probability where every event has equal chances of occuring.

x = random.uniform(size=(2, 3))

print(x)

sns.distplot(random.uniform(size=1000), hist=False)

plt.show()
```

```python
# Exponential Distribution
# Exponential distribution is used for describing time till next event e.g. failure/success etc.

x = random.exponential(scale=2, size=(2, 3)) # inverse of rate ( see lam in poisson distribution ) defaults to 1.0.

print(x)

sns.distplot(random.exponential(size=1000), hist=False)

plt.show()
```

```python
# Chi Square Distribution
# Chi Square distribution is used as a basis to verify the hypothesis.

x = random.chisquare(df=2, size=(2, 3)) # df : degree of freedom

print(x)

sns.distplot(random.chisquare(df=1, size=1000), hist=False)

plt.show()
```

### uFunc
ufuncs stands for "Universal Functions" and they are NumPy functions that operates on the ndarray object.

```python
print(type(np.add))
print(type(np.concatenate))
```

```python
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([20, 21, 22, 23, 24, 25])

newarr = np.multiply(arr1, arr2)

print(newarr)


newarr = np.power(arr1, arr2)

print(newarr)

newarr = np.divmod(arr1, arr2) # quotient and mode

print(newarr)

arr = np.array([-1, -2, 1, 2, 3, -4])

newarr = np.absolute(arr)

print(newarr)
```

```python
arr = np.trunc([-3.1666, 3.6667])

print(arr)

arr = np.fix([-3.1666, 3.6667])

print(arr)

arr = np.floor([-3.1666, 3.6667])

print(arr)

arr = np.ceil([-3.1666, 3.6667])

print(arr)


arr = np.around(3.1666, 2)

print(arr)
```

```python
arr = np.arange(1, 10)

print(np.log2(arr))

print(np.log10(arr))

print(np.log(arr)) # base e ( log e = ln )
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([1, 2, 3])

newarr = np.add(arr1, arr2)

print(newarr)

newarr = np.sum([arr1, arr2])

print(newarr)

newarr = np.sum([arr1, arr2], axis=1)

print(newarr)

newarr = np.cumsum(arr1)

print(newarr)
```

```python
arr = np.array([1, 2, 3, 4])

x = np.prod(arr)

print(x)

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])

x = np.prod([arr1, arr2])

print(x) # 1*2*3*4*5*6*7*8 = 40320

newarr = np.prod([arr1, arr2], axis=1)

print(newarr)

newarr = np.cumprod(arr1)

print(newarr)
```

```python
arr = np.array([10, 15, 25, 5])

newarr = np.diff(arr)

print(newarr)

newarr = np.diff(arr, n=2)

print(newarr) # 15-10=5, 25-15=10, 5-25=-20 AND 10-5=5, -20-10=-30
```

```python
num1 = 4
num2 = 6

x = np.lcm(num1, num2)

print(x)

arr = np.array([3, 6, 9])

x = np.lcm.reduce(arr)

print(x)

arr = np.arange(1, 11)

x = np.lcm.reduce(arr)

print(x)

num1 = 6
num2 = 9

x = np.gcd(num1, num2)

print(x)

arr = np.array([20, 8, 32, 36, 16])

x = np.gcd.reduce(arr)

print(x)
```

```python
x = np.sin(np.pi/2)

print(x)

arr = np.array([np.pi/2, np.pi/3, np.pi/4, np.pi/5])

x = np.sin(arr)

print(x)

arr = np.array([90, 180, 270, 360])

x = np.deg2rad(arr)

print(x)

arr = np.array([np.pi/2, np.pi, 1.5*np.pi, 2*np.pi])

x = np.rad2deg(arr)

print(x)

x = np.arcsin(1.0)

print(x)

arr = np.array([1, -1, 0.1])

x = np.arcsin(arr)

print(x)

base = 3
perp = 4

x = np.hypot(base, perp)

print(x)

x = np.sinh(np.pi/2)

print(x)

arr = np.array([np.pi/2, np.pi/3, np.pi/4, np.pi/5])

x = np.cosh(arr)

print(x)

x = np.arcsinh(1.0)

print(x)

arr = np.array([0.1, 0.2, 0.5])

x = np.arctanh(arr)

print(x)
```

```python
arr = np.array([1, 1, 1, 2, 3, 4, 5, 5, 6, 7])

x = np.unique(arr)

print(x)

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])

newarr = np.union1d(arr1, arr2)

print(newarr)

newarr = np.intersect1d(arr1, arr2, assume_unique=True)

print(newarr)

newarr = np.setdiff1d(arr1, arr2, assume_unique=True)

print(newarr)

newarr = np.setxor1d(arr1, arr2, assume_unique=True)

print(newarr)

# assume_unique, which if set to True can speed up computation. It should always be set to True when dealing with sets.
```
